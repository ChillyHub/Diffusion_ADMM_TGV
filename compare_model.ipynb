```python
import os
import analyzing_tools as at
```

```python
config_name = 'AAPM_256_ncsnpp_continuous'
problem = 'sparseview_CT_ADMM_TV_total'

vo_file_name = 'volume_198.npy'
gt_file_name = 'ground_truth_198.npy'
```

# Sparse 8 view

```python
# view 8

volume_8views = []
method_8views = []

try:
    method = 'TV'
    root = f'./results/{config_name}/{problem}/{method}/n_view{8}/rho_{10}-lambda_{0.04}/volume/'
    volume_TV_8views, ground_truth = at.load_volume(root, vo_file_name, gt_file_name)
    volume_8views.append(volume_TV_8views)
    method_8views.append(method)
except:
    pass

try:
    method = 'TGV_3'
    root = f'./results/{config_name}/{problem}/{method}/n_view{8}/rho0_{5}-rho1_{5}-lambda_{0.04}-alpha0_{2.5}-alpha_1_{2.5}/volume/'
    volume_TGV_3_8views, ground_truth = at.load_volume(root, vo_file_name, gt_file_name)
    volume_8views.append(volume_TGV_3_8views)
    method_8views.append(method)
except:
    pass

try:
    method = 'TGV_2'
    root = f'./results/{config_name}/{problem}/{method}/n_view{8}/rho0_{5}-rho1_{5}-lambda_{0.04}-alpha0_{2.5}-alpha_1_{2.5}/volume/'
    volume_TGV_2_8views, ground_truth = at.load_volume(root, vo_file_name, gt_file_name)
    volume_8views.append(volume_TGV_2_8views)
    method_8views.append(method)
except:
    pass

try:
    method = 'TGV_1'
    root = f'./results/{config_name}/{problem}/{method}/n_view{8}/rho0_{5}-rho1_{5}-lambda_{0.04}-alpha0_{2.5}-alpha_1_{2.5}/volume/'
    volume_TGV_1_8views, ground_truth = at.load_volume(root, vo_file_name, gt_file_name)
    volume_8views.append(volume_TGV_1_8views)
    method_8views.append(method)
except:
    pass
```

```python
dict_8views_list = []
for i in range(len(volume_8views)):
    reslt_8view = at.extract_slices_and_evaluate(volume_8views[i], ground_truth)
    dict_8views_list.append(at.flatten_nested_dict(reslt_8view))

data_8views_df = at.flatten_dicts_to_df(dict_8views_list, method_8views)

at.table_comparison(data_8views_df, 'PSNR, SSIM')
```

```python
at.plot_comparison(volume_8views, method_8views, ground_truth, inset=True)
```
